```python
import gradio as gr
import json
import time
import os
import random

def load_conversations(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file {file_path} does not exist.")
    conversations = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            convo = json.loads(line.strip())
            conversations.append(convo)
    return conversations

negotiation_chat_file = "negotiation_finetune_data_base.jsonl"
conversations = load_conversations(negotiation_chat_file)

# Assign IDs if none exist
for i, convo in enumerate(conversations):
    if 'ConversationID' not in convo:
        convo['ConversationID'] = f"conversation_{i}"

used_conversation_ids = set()
stop_simulation = False
current_conversation_id = None

def get_random_unused_conversation():
    global used_conversation_ids

    # Filter only those conversations that have 'messages' and are not empty
    valid_conversations = [
        convo for convo in conversations
        if 'messages' in convo and isinstance(convo['messages'], list) and convo['messages']
    ]

    # If all have been used, reset
    if len(used_conversation_ids) == len(valid_conversations):
        used_conversation_ids.clear()

    # Filter out used ones
    unused_conversations = [
        convo for convo in valid_conversations
        if convo['ConversationID'] not in used_conversation_ids
    ]

    if not unused_conversations:
        return None

    selected_conversation = random.choice(unused_conversations)
    return selected_conversation

def simulate_conversation():
    global stop_simulation, current_conversation_id, used_conversation_ids
    stop_simulation = False
    
    selected_conversation = get_random_unused_conversation()
    if not selected_conversation:
        yield "<div>No unused valid conversations available.</div>"
        return
    
    current_conversation_id = selected_conversation.get('ConversationID')
    used_conversation_ids.add(current_conversation_id)
    messages = selected_conversation.get('messages', [])

    if not messages:
        yield f"<div>No messages found for Conversation ID: {current_conversation_id}</div>"
        return

    chat_html = f"<div>Simulation ID: {current_conversation_id}</div><br>"
    for msg in messages:
        if stop_simulation:
            yield gr.update(value="<div>Simulation interrupted. Click 'Start Simulation' to start a new one.</div>")
            return

        # Format messages based on role
        if msg['role'] == "user":
            # User messages on the left
            chat_html += f"""
            <div style="text-align: left; margin: 10px;">
                <div style="display: inline-block; padding: 10px; border-radius: 15px;
                            background-color: #d1f4ff; max-width: 60%; font-size: 18px;color: black;">
                    <b>{msg['role'].capitalize()}</b>: {msg['content']}
                </div>
            </div>
            """
        elif msg['role'] == "assistant":
            # Assistant messages on the right
            chat_html += f"""
            <div style="text-align: right; margin: 10px;">
                <div style="display: inline-block; padding: 10px; border-radius: 15px;
                            background-color: #e6ffe6; max-width: 60%; font-size: 18px;color: black;">
                    <b>{msg['role'].capitalize()}</b>: {msg['content']}
                </div>
            </div>
            """
        elif msg['role'] == "system":
            # System messages center-aligned
            chat_html += f"""
            <div style="text-align: center; margin: 20px; font-size: 16px;color: black;">
                <i><b>{msg['role'].capitalize()}</b>: {msg['content']}</i>
            </div>
            """

        yield gr.update(value=chat_html)
        time.sleep(2.5)  # simulate typing delay

    chat_html += """
    <div style="text-align: center; margin: 20px; font-size: 14px;">
        Negotiation simulation ended.
    </div>
    """
    yield gr.update(value=chat_html)

def stop_current_simulation():
    global stop_simulation
    stop_simulation = True
    return "<div>Simulation interrupted. Click 'Start Simulation' to start a new one.</div>"

with gr.Blocks() as demo:
    gr.Markdown("## Self-Play Negotiation Simulation Demo")
    gr.Markdown("Click 'Start Simulation' to begin the simulation or 'Stop Simulation' to interrupt it.")
    
    start_button = gr.Button("Start Simulation")
    stop_button = gr.Button("Stop Simulation")
    output_box = gr.HTML(label="Live Chat Simulation")

    start_button.click(simulate_conversation, inputs=None, outputs=output_box)
    stop_button.click(stop_current_simulation, inputs=None, outputs=output_box)

demo.launch()
```

```python
import gradio as gr
import json
import time
import os
import random

def load_conversations(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file {file_path} does not exist.")
    conversations = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            convo = json.loads(line.strip())
            conversations.append(convo)
    return conversations

negotiation_chat_file = "negotiation_finetune_data_base.jsonl"
conversations = load_conversations(negotiation_chat_file)

# Assign IDs if none exist
for i, convo in enumerate(conversations):
    if 'ConversationID' not in convo:
        convo['ConversationID'] = f"conversation_{i}"

used_conversation_ids = set()
stop_simulation = False
current_conversation_id = None

def get_random_unused_conversation():
    global used_conversation_ids

    # Filter only those conversations that have 'messages' and are not empty
    valid_conversations = [
        convo for convo in conversations
        if 'messages' in convo and isinstance(convo['messages'], list) and convo['messages']
    ]

    # If all have been used, reset
    if len(used_conversation_ids) == len(valid_conversations):
        used_conversation_ids.clear()

    # Filter out used ones
    unused_conversations = [
        convo for convo in valid_conversations
        if convo['ConversationID'] not in used_conversation_ids
    ]

    if not unused_conversations:
        return None

    selected_conversation = random.choice(unused_conversations)
    return selected_conversation

def simulate_conversation():
    global stop_simulation, current_conversation_id, used_conversation_ids
    stop_simulation = False
    
    selected_conversation = get_random_unused_conversation()
    if not selected_conversation:
        yield "<div>No unused valid conversations available.</div>"
        return
    
    current_conversation_id = selected_conversation.get('ConversationID')
    used_conversation_ids.add(current_conversation_id)
    messages = selected_conversation.get('messages', [])

    if not messages:
        yield f"<div>No messages found for Conversation ID: {current_conversation_id}</div>"
        return

    # Role mapping
    role_map = {
        "assistant": "HR Manager",
        "user": "Employee",
        "system": "System"
    }

    chat_html = f"<div>Simulation ID: {current_conversation_id}</div><br>"
    for msg in messages:
        if stop_simulation:
            yield gr.update(value="<div>Simulation interrupted. Click 'Start Simulation' to start a new one.</div>")
            return

        # Determine display role name
        display_role = role_map.get(msg['role'], msg['role'].capitalize())

        if msg['role'] == "user":
            # Employee messages on the left
            chat_html += f"""
            <div style="text-align: left; margin: 10px;">
                <div style="display: inline-block; padding: 10px; border-radius: 15px;
                            background-color: #d1f4ff; max-width: 60%; font-size: 18px; color: black;">
                    <b style="color: black;">{display_role}</b>: <span style="color: black;">{msg['content']}</span>
                </div>
            </div>
            """
        elif msg['role'] == "assistant":
            # HR Manager messages on the right
            chat_html += f"""
            <div style="text-align: right; margin: 10px;">
                <div style="display: inline-block; padding: 10px; border-radius: 15px;
                            background-color: #e6ffe6; max-width: 60%; font-size: 18px; color: black;">
                    <b style="color: black;">{display_role}</b>: <span style="color: black;">{msg['content']}</span>
                </div>
            </div>
            """
        elif msg['role'] == "system":
            # System messages center-aligned
            chat_html += f"""
            <div style="text-align: center; margin: 20px; font-size: 16px;">
                <i><b style="color: white;">{display_role}</b>: <span style="color: white;">{msg['content']}</span></i>
            </div>
            """



        yield gr.update(value=chat_html)
        time.sleep(2.5)  # simulate typing delay

    chat_html += """
    <div style="text-align: center; margin: 20px; font-size: 14px;">
        Negotiation simulation ended.
    </div>
    """
    yield gr.update(value=chat_html)

def stop_current_simulation():
    global stop_simulation
    stop_simulation = True
    return "<div>Simulation interrupted. Click 'Start Simulation' to start a new one.</div>"

with gr.Blocks() as demo:
    gr.Markdown("## Self-Play Negotiation Simulation Demo")
    gr.Markdown("Click 'Start Simulation' to begin the simulation or 'Stop Simulation' to interrupt it.")
    
    start_button = gr.Button("Start Simulation")
    stop_button = gr.Button("Stop Simulation")
    output_box = gr.HTML(label="Live Chat Simulation")

    start_button.click(simulate_conversation, inputs=None, outputs=output_box)
    stop_button.click(stop_current_simulation, inputs=None, outputs=output_box)

demo.launch()
```
